```python
#eda
import pandas as pd
import numpy as np

import scipy.stats as st

import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams

#import data
train = pd.read_csv('../dataset/df_tot.csv')
target = 'Live_Starts'
IDcol = 'ID'
predictors = [x for x in train.columns if x not in [target, IDcol]]
```

```python
#correlations
from scipy.stats import pearsonr

cpreds_ha = [x for x in predictors if x in ['Attendance', 'Capacity', 'sellout',
                                         'Num_Secondary_Transactions', 'Avg_Markup',
                                         'season_tickets', 'group_tickets', 'single_game_tickets',
                                         'dma_home', 'dma_away',
                    
                                        'Home_Followers_Facebook', 'Home_Engagements_Facebook', 'Home_Impressions_Facebook',
                                        'Away_Followers_Facebook', 'Away_Engagements_Facebook', 'Away_Impressions_Facebook',
                                        'Home_Followers_Instagram', 'Home_Engagements_Instagram',
                                        'Away_Followers_Instagram', 'Away_Engagements_Instagram',
                                        'Home_Followers_Snapchat', 
                                        'Away_Followers_Snapchat',
                                        'Home_Followers_Weibo', 
                                        'Away_Followers_Weibo',
                                        'Home_Followers_Twitter', 'Home_Engagements_Twitter', 'Home_Impressions_Twitter',
                                        'Away_Followers_Twitter', 'Away_Engagements_Twitter', 'Away_Impressions_Twitter']]



cpreds = [x for x in predictors if x in ['Attendance', 'Capacity', 'sellout',
                                         'Num_Secondary_Transactions', 'Avg_Markup',
                                         'season_tickets', 'group_tickets', 'single_game_tickets',
                                         
                                         'gt_tot', 'gt2_tot', 'gtq_tot', 'gt2q_tot',
                                         
                                         'dma_tot','fbf_tot', 'twf_tot', 'igf_tot', 'scf_tot', 'wbf_tot', 
                                         'fbe_tot', 'twe_tot', 'ige_tot', 'fbi_tot', 'twi_tot']]

correlations = {}
for f in cpreds:
    data_temp = train[[f,target]]
    x1 = data_temp[f].values
    x2 = data_temp[target].values
    key = f + ' vs ' + target
    correlations[key] = pearsonr(x1,x2)[0]

corr = pd.DataFrame(correlations, index=['Value']).T
corr.loc[corr['Value'].abs().sort_values(ascending=False).index]
```
